```python
data['Deck'] = data['Cabin'].str[0]
```

```python
print(data['Deck'].isnull().sum())
```

```python
data['Deck'] = data.groupby('Pclass')['Deck'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
data['Has_Cabin'] = data['Cabin'].notnull().astype(int)
```

```python
data = data.drop(columns=['Cabin'])
```

```python
data['Embarked'].count()
```

```python
data['Embarked'].value_counts()
```

```python
data['Embarked'] = data['Embarked'].fillna('S')
```

### Data Preprocessing Complet Here

```python
x = data.drop(['Survived'], axis=1)

y = data['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42, stratify = y)
```

```python
x_train.shape, x_test.shape
```

```python
x_train.dtypes
```

```python
x_train.head()
```

```python
import category_encoders as ce
```

```python
encoder = ce.OrdinalEncoder(cols=['PassengerId', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Embarked', 'Deck', 'Has_Cabin'])


x_train = encoder.fit_transform(x_train)

x_test = encoder.transform(x_test)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# instantiate the DecisionTreeClassifier model with criterion gini index

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# fit the model
clf_gini.fit(x_train, y_train)
```

```python
y_pred_gini = clf_gini.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

```python
y_pred_train_gini = clf_gini.predict(x_train)

y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

```python
print('Training set score: {:.4f}'.format(clf_gini.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(x_test, y_test)))
```

```python
# instantiate the DecisionTreeClassifier model with criterion entropy

clf_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)


# fit the model
clf_en.fit(x_train, y_train)
```

```python
y_pred_en = clf_en.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion entropy: {0:0.4f}'. format(accuracy_score(y_test, y_pred_en)))
```

```python
y_pred_train_en = clf_en.predict(x_train)

y_pred_train_en
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_en)))
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(clf_en.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(clf_en.score(x_test, y_test)))
```

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_en))
```
